```python
from flask import Flask, jsonify
import subprocess, json, os

app = Flask(__name__)

BASE_DIR = os.path.dirname(os.path.abspath(__file__))
SCRIPT   = os.path.join(BASE_DIR, "spx_signal_v91.py")
BT_JSON  = os.path.join(BASE_DIR, "output", "backtest_result.json")

def run_script(args):
    """调用现有 py，返回 stdout 的 JSON。"""
    cmd = ["python", SCRIPT] + args
    out = subprocess.check_output(cmd, cwd=BASE_DIR).decode("utf-8")
    return json.loads(out)

@app.get("/api/live")
def api_live():
    # 当前月信号
    data = run_script(["--json"])
    return jsonify(data)

@app.get("/api/backtest")
def api_backtest():
    # 回测结果（含 summary / yearly / monthly_pnl）
    data = run_script(["--backtest", "--json"])
    # 可选：也从文件读，例如：
    # with open(BT_JSON) as f:
    #     data = json.load(f)
    return jsonify(data)

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=8000, debug=True)
```
